```python
def fibonacci_series():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

def nth_fibonacci(n):
    fib_gen = fibonacci_series
    for _ in range(n):
        yield next(fib_gen)
    return fib_gen

n = 10
nth_fib = nth_fibonacci(n)
# print(f"{next(nth_fibonacci(n))}")
print(f"The {n}th Fibonacci number is: {nth_fib}")
```

```python
class Product:
    def __init__(self, id, name, price):
        self.id = id
        self.name = name
        self.price = price

product_list = [
    Product(1, "Sony mobile", 25000),
    Product(2, "Lenovo mobile", 15000),
    Product(3, "Nokia mobile", 10000),
    Product(4, "Samsung mobile", 40000),
    Product(5, "Real Me", 50000)
]

def search_products_less_than_price(products, max_price):
    return [product for product in products if product.price < max_price]

def count_products_less_than_price(products, max_price):
    return sum(1 for product in products if product.price < max_price)

def display_product_list(products):
    print("Product List:")
    for product in products:
        print(f"{product.name} - ${product.price}")
    print()

def sum_of_product_prices(products):
    return sum(product.price for product in products)

def display_summary_statistics(products):
    total_products = len(products)
    total_price = sum_of_product_prices(products)
    print("Summary Statistics of Products:")
    print(f"Total number of products: {total_products}")
    print(f"Total sum of all product prices: ${total_price}")
    print()

def max_price_product(products):
    return max(products, key=lambda product: product.price)

def min_price_product(products):
    return min(products, key=lambda product: product.price)

def display_product_name_price(products):
    print("Product Name along with Price:")
    for product in products:
        print(f"{product.name} - ${product.price}")
    print()

def display_product_names(products):
    print("List of Product Names:")
    for product in products:
        print(product.name)


max_price = 30000

less_than_price = search_products_less_than_price(product_list, max_price)
print(f"Products less than {max_price} price:")
for product in less_than_price:
    print(f"{product.name} - ${product.price}")
print()

count_price = count_products_less_than_price(product_list, max_price)
print(f"Number of products with price less than {max_price}: {count_price}")
print()

display_product_list(product_list)

total_price = sum_of_product_prices(product_list)
print(f"Sum of all product prices: ${total_price}")
print()

display_summary_statistics(product_list)

max_price_prod = max_price_product(product_list)
print("Product with max price:")
print(f"{max_price_prod.name} - ${max_price_prod.price}")
print()

min_price_prod = min_price_product(product_list)
print("Product with min price:")
print(f"{min_price_prod.name} - ${min_price_prod.price}")
print()

display_product_name_price(product_list)
print()

display_product_names(product_list)
```

```python
class InvalidNameException(Exception):
    pass

class InvalidJobProfileException(Exception):
    pass

class InvalidAgeException(Exception):
    pass


class Applicant:
    def __init__(self, name, job_profile, age):
        self.name = name
        self.job_profile = job_profile
        self.age = age

class Validator:
    @staticmethod
    def validate_name(name):
        if not name:
            return False
        return True
    
    @staticmethod
    def validate_job_profile(job_profile):
        valid_profiles = ['Associate', 'Clerk', 'Executive', 'Officer']
        if job_profile.lower() in map(str.lower, valid_profiles):
            return True
        return False
    
    @staticmethod
    def validate_age(age):
        if 18 <= age <= 30:
            return True
        return False
    
    def validate(self, applicant):
        if not self.validate_name(applicant.name):
            raise InvalidNameException("Invalid Name")
        
        if not self.validate_job_profile(applicant.job_profile):
            raise InvalidJobProfileException("Invalid Job Profile")
        
        if not self.validate_age(applicant.age):
            raise InvalidAgeException("Invalid Age")
        
        return True

def main():
    try:
        # Sample applicant data
        applicant = Applicant("John Doe", "Executive", 25)
        
        # Create a Validator instance
        validator = Validator()
        
        # Validate the applicant details
        if validator.validate(applicant):
            print("Application submitted successfully!")
    
    except InvalidNameException as e:
        print(e)
    
    except InvalidJobProfileException as e:
        print(e)
    
    except InvalidAgeException as e:
        print(e)

# Run the main function
if __name__ == "__main__":
    main()
```
